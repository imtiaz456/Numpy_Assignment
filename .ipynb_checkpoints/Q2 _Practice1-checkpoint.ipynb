# Handling missing values 
  - filling missing values
  - Dropping missing values

```python
data.info()
```

```python
data.year.fillna(data.year.mode(),inplace=True)
```

```python
data.year.isnull().sum()
```

```python
data.columns
```

```python
data[data.alma_mater.isnull()]
```

```python
data[data.group.isnull()]
```

```python
data.fillna({
    'year':data.year.mode(),
    'missions':data.missions.mode(),
    'birthplace':data.birthplace.mode()
})
```
